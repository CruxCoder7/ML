```python
import torch 
from torch.utils.data import TensorDataset, DataLoader
import torch.nn as nn
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
torch.set_default_dtype(torch.float64)
```

```python
df = pd.read_csv(r"C:\Users\91755\Downloads\Bank_Personal_Loan_Modelling.csv")
```

```python
df = df.drop(['ID'],axis=1)
x = df.drop(['Personal Loan'],axis=1).values
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
x[:5]
```

```python
y = df['Personal Loan'].values 
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=10,test_size=0.3)
```

```python
x_train = np.array(x_train)
x_train = torch.from_numpy(x_train)
x_train = x_train.double()
x_train.shape
```

```python
y_train = np.array(y_train)
y_train = torch.from_numpy(y_train)
y_train = y_train.double()
```

```python
tr = TensorDataset(x_train,y_train)
tr[0]
```

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(12,10),
            nn.ReLU(),
            nn.Linear(10,5),
            nn.Sigmoid(),
            nn.Linear(5,1)
        )
    
    def forward(self,x):
        logits = self.linear_relu_stack(x)
        return logits
```

```python
model = Net()
model
```

```python
loss = nn.MSELoss()
optim = torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
def train(model,epoch,data):
    for i in range(epoch):
        for a,b in data:
            pr = model(a)
            b = b.unsqueeze(0)
            l = loss(pr,b)
            l.backward()
            optim.step()
            optim.zero_grad()
            if i%10 == 0:
                print(f"Epoch: {i+1}, Error: {l}")
```

```python
train(model,20,tr)
```

```python
x_test
```

```python
pred = model(torch.from_numpy(x_test))
```

```python
pred = pred > 0.5
```

```python
lc = LabelEncoder()
pred = lc.fit_transform(pred)
```

```python
accuracy_score(y_test,pred)
```
