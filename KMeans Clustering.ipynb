```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\91755\Downloads\data_new.csv")
```

```python
df.head()
```

```python
data = df.iloc[:,2:].values
data
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i,random_state=0)
    km.fit(data)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters=5,random_state=0)
km.fit(data)
```

```python
y = km.predict(data)
y 
```

```python
import seaborn as sns
sns.set(style='darkgrid')
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[y == 0,0],data[y == 0,1],data[y == 0,2], s = 40 , color = 'blue', label = "cluster 0")
ax.scatter(data[y == 1,0],data[y == 1,1],data[y == 1,2], s = 40 , color = 'orange', label = "cluster 1")
ax.scatter(data[y == 2,0],data[y == 2,1],data[y == 2,2], s = 40 , color = 'green', label = "cluster 2")
ax.scatter(data[y == 3,0],data[y == 3,1],data[y == 3,2], s = 40 , color = 'red', label = "cluster 3")
ax.scatter(data[y == 4,0],data[y == 4,1],data[y== 4,2], s = 40 , color = 'purple', label = "cluster 4")
ax.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],km.cluster_centers_[:,2],color='black',s=150,label='centroids')
ax.set_xlabel('Age ')
ax.set_ylabel('Anual Income')
ax.set_zlabel('Spending Score')
ax.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris

data = load_iris()
data
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i,random_state=0)
    km.fit(df)
    wcss.append(km.inertia_)
```
