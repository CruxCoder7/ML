```python
df = df.drop(['ID'],axis=1)
```

```python
x = df.drop(['Personal Loan'],axis=1).values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
y = df['Personal Loan'].values 
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=10,test_size=0.3)
```

```python
import tensorflow as tf 
```

```python
tf.random.set_seed(40)
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=[12]))
model.add(tf.keras.layers.Dense(8,activation='relu'))
model.add(tf.keras.layers.Dense(4,activation='relu'))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='BinaryCrossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs=50)
```

```python
model.evaluate(x_test,y_test)
```

```python
pred = model.predict(x_test)
```

```python
y_pred = pred > 0.5 
y_pred
```

```python
y_pred 
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)
```

## PyTorch

```python
import torch 
import numpy as np
```

```python
torch.set_default_dtype(torch.float64)
```

```python
x_train2 = np.array(x_train)
x_train2 = torch.from_numpy(x_train2)
x_train2 = x_train2.double()
```

```python
y_train2 = np.array(y_train)
y_train2 = torch.from_numpy(y_train2)
y_train2 = y_train2.double()
```

```python
from torch.utils.data import TensorDataset 
from torch.utils.data import DataLoader 
from torch import nn
```

```python
tr = TensorDataset(x_train2,y_train2)
```

```python
tr[0]
```

```python
x_train_dl = DataLoader(tr,batch_size=10,shuffle=True)
```

```python
class NN(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(12,10),
            nn.ReLU(),
            nn.Linear(10,5),
            nn.Sigmoid(),
            nn.Linear(5,1)
        )
    def forward(self,x):
        logits = self.linear_relu_stack(x) 
        return logits
```

```python
m = NN()
m
```

```python
pr = model(x_train)
```

```python
pred = pr > 0.5 
```

```python
lc = LabelEncoder()
pred = lc.fit_transform(pred)
```

```python
pred
```

```python
accuracy_score(y_train,pred)
```

```python
loss = nn.MSELoss()
```

```python
optim = torch.optim.SGD(m.parameters(),lr=0.01)
```

```python
def train(m,ep,d):
    for i in range(ep):
        for a,b in d: 
            pr = m(a)
            b = b.unsqueeze(0)
            l = loss(pr,b)
            l.backward()
            optim.step()
            optim.zero_grad()
            if i % 10 == 0:
                print(f"Epoch: {i+1}, Error: {l}")
```

```python
train(m,50,tr)
```

```python
pred = m(torch.from_numpy(x_test))
```

```python
pred = pred > 0.5 
```

```python
pred = lc.fit_transform(pred.flatten())
```

```python
accuracy_score(y_test,pred)
```
